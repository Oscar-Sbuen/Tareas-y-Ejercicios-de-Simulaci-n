# Elegir al azar una ciudad de la Lista despues de hacer una mezcla

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
Lista = ['Rome','New York','London','Berlin','Moskov', 'Los Angeles','Paris','Madrid','Tokio','Toronto']
random.shuffle(Lista)
print('La lista mezclada es:',Lista,'\t \t')
print('\t')
print('La ciudad de esa mezcla es - ', random.choice(Lista))
```

# Elegir al azar tres ciudades despues de una mezcla.

```python
for i in range(3):
    print('Las tres ciudades al azar son -',i+1,random.choice(Lista))
```

# Ejercicio Mtodo de las doce normales



```python
# Utilizando la funcin random
#mu=0
#sigma^2=1
lis=[]
n=100000
cont=0
for i in range(n):
    for i in range(12):
        cont=cont+random.random()
    lis.append(cont-6)
    cont=0

plt.hist(lis,bins=200,edgecolor="black")
plt.title('Las doce uniformes')
plt.show()



nums = []
mu = 100
sigma = 50

for i in range(n):
    temp = random.gauss(mu, sigma)
    nums.append(temp)

plt.hist(nums, bins = 200,edgecolor="gold")
plt.title('Utilizando la funcin Gauss')
plt.show()
```
