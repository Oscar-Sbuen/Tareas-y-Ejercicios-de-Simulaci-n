# ValorizaciÃ³n de Opciones Europeas.

####  OpciÃ³n Call


```python
import numpy as np
import scipy.stats as st

def BS_Vanilla_Call(S, K, T, r, sigma):
    #S: precio spot
    #K: precio strike
    #T: tiempo de vencimiento
    #r: tasa de interes
    #sigma: volatilidad del activo
    
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - (sigma * np.sqrt(T))
    C = (S * st.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * st.norm.cdf(d2, 0.0, 1.0))
    return C

#Probando la funciÃ³n de Python
a = BS_Vanilla_Call(90, 100, 1, 0.05, 0.25)

print("Precio Call Option: {:0.6f}".format(a))
```

### OpciÃ³n Put

```python
def BS_Vanilla_Put(S, K, T, r, sigma):
    #S: precio spot
    #K: precio strike
    #T: tiempo de vencimiento
    #r: tasa de interes
    #sigma: volatilidad del activo
    
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - (sigma * np.sqrt(T))
    P = K * np.exp(-r * T) * st.norm.cdf(-d2, 0.0, 1.0) - S * st.norm.cdf(-d1, 0.0, 1.0) 
    return P

#Probando la funciÃ³n de Python
a = BS_Vanilla_Put(90, 100, 1, 0.05, 0.25)

print("Precio Put Option: {:0.6f}".format(a))
```

## Valorizacion de Opciones con QuantLib.
Bueno ahora les presentarÃ© QuantLib. QuantLib es una libreria de C, diseÃ±ada para la valorizacion de instrumentos financieros (opciones, forwardas, swaps y bonos). Como la libreria esta diseÃ±ada en C la nomenclatura es algo complicada, pero las ventajas de esta libreria es que tiene desarrollada la valorizacion de muchos instrumentos financieros bajo diferentes metodos (arboles binomiales, forma analitica, simulacion de montecarlo entre otras), tambien las griegas de opciones y duracion y convexidad para bonos. Entre las desventajas estan que la documentaciÃ³n es algo confusa para los que no saben C o no estan familiarizados con algun lenguaje de programaciÃ³n.

Ahora sin mas preambulos vamos a crear una opciÃ³n call con QuantLibPython.

```python
T = ql.Date(15, 2, 2019)
S = ql.SimpleQuote(90)
K = 100
sigma = 0.25 # the historical vols for a year
div_rate =  0.00
option_type = ql.Option.Call

r = 0.05
day_count = ql.Actual365Fixed()
calendar = ql.UnitedStates()

t_0 = T - 365
ql.Settings.instance().evaluationDate = t_0

# Construimos la opciÃ³n
payoff = ql.PlainVanillaPayoff(option_type, K)
exercise = ql.EuropeanExercise(T)
call_option = ql.VanillaOption(payoff, exercise)

# Construimos el proceso para valorizar la opciÃ³n
# En este caso usaremos un proceso estocastico de Black Scholes Merton
spot_handle = ql.QuoteHandle(S)
flat_ts = ql.YieldTermStructureHandle(ql.FlatForward(0, calendar, r, day_count))
div_yield = ql.YieldTermStructureHandle(ql.FlatForward(0, calendar, div_rate, day_count))
flat_vol_ts = ql.BlackVolTermStructureHandle(ql.BlackConstantVol(0, calendar, sigma, day_count))
bsm_process = ql.BlackScholesMertonProcess(spot_handle, 
                                           div_yield, 
                                           flat_ts, 
                                           flat_vol_ts)

# AÃ±adimos el proceso a la opcion
call_option.setPricingEngine(ql.AnalyticEuropeanEngine(bsm_process))

bs_price1 = call_option.NPV()

bs_price = BS_Vanilla_Call(90, 100, 1, 0.05, 0.25)

print("Precio Call Option Python Puro : {:0.6f}".format(bs_price))
print("Precio Call Option con QuantLib: {:0.6f}".format(bs_price1))
```

# ValorizaciÃ³n de Opciones Americanas.
## OpciÃ³n Call
Las opciones americanas son opciones que a diferencia de las europeas, pueden ser ejercidas en cualquier momento del tiempo. No tienen una formula explicita para el calculo de su precio, sin embargo se puede calcular sus precios con varios metodos como los metodos analiticos, arboles binomiales, diferencias finitas y simulacion de montecarlo:
Ahora vamos a probar los diferentes metodos para valorizar una opciÃ³n americana yeeee =D.

```python
steps = 100
S.setValue(90)
exercise_am = ql.AmericanExercise(t_0, T)
call_option_am = ql.VanillaOption(payoff, exercise_am)
```

# Usando los arboles binomiales.

```python
binomials = ['jr','crr','eqp','trigeorgis','tian','lr',]
for i in binomials:
    binomial_engine = ql.BinomialVanillaEngine(bsm_process, i, steps)
    call_option_am.setPricingEngine(binomial_engine)
    print("Binomial ("+ str(i) +") : {:0.6f}".format(call_option_am.NPV()))
```

# Usando la simulacion de MonteCarlo.

```python
montecarlo_engine = ql.MCAmericanEngine(bsm_process, 'PseudoRandom', 
                                        timeSteps = steps,
                                        requiredSamples = 1000,
                                        requiredTolerance = 0.01,
                                        seed = 1,
                                        seedCalibration = 1)

call_option_am.setPricingEngine(montecarlo_engine)
print("MonteCarlo: {:0.6f}".format(call_option_am.NPV()))
```
