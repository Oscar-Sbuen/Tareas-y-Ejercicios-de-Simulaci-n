b)Usar la función random() para simular 1000 volados 


c)Determinar de b) la probabilidad de que ocurra "sol" en un volado.

```python
#by:Oscar
contador=0
n=1000
for i in range(n):
    x=random.random()
    if x<=0.5:
        print("Aguíla")
    else:
        print("Sol")
        contador=contador+1
print("La probabilidad de que salga sol es:", contador/n)
        
```

# Tarea 1
Estimar el valor de pi usando la función random.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#d=distancia entre el origen y el punto

x = np.linspace(0, 1, 100)
y = np.sqrt(1 - x**2)               # Ecuación del círculo en el primer cuadrante 
conta_1=0
conta_2=0

for i in range(1000000):
    x_1 = random.random()
    y_1 = random.random()
    d=math.sqrt((x_1)**2+(y_1)**2)
    if d<=1:
        conta_1+=1
    conta_2+=1
pi=(conta_1/conta_2)*4
print("La simulación da:",'%4.3f' % pi)


fig = plt.figure(1, figsize=(8, 8))        # Aquí fijamos el formato de la gráfica para que sea cuadrada.
plt.plot(x, y, color="red", markersize=1) 
plt.scatter(x_1, y_1, s=20, c="blue")
plt.title("Estimación de $\pi$")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```
