```python
from random import random
import matplotlib.pyplot as plt
N=1000000
n=0
lx=[]
ly=[]
for i in range(N):
    x=random()
    y=random()
    if x**2+y**2<1:
        n=n+1
    if i%10000==0 and i>10:
        est=4*n/(i+1)
        lx.append(i+1)
        ly.append(est)
print("La simulaci√≥n da:", "%4.4f"%est)
plt.plot(lx,ly)
plt.show()
```
